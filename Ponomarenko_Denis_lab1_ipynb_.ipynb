Linear regression: y = a*x + b

```python
LEARNING_RATE = 0.01
PRECISION = 0.000001
```

```python
def mean_squared_error(y_true, y_predicted):
    return np.mean((y_true - y_predicted)**2)

def mean_squared_error_d(y_true, y_predicted):
    return 2*(y_predicted - y_true)
```

```python
b = 0
a = np.random.normal(0, 1)
```

```python
(a, b)
```

dy/da = x, dy/db = 1

```python
change = np.inf
cur_loss = 0
e = 0
while change > PRECISION:
    y_pred = a * X_train + b
    loss = mean_squared_error(y_train, y_pred)
    change = np.abs(loss - cur_loss)
    cur_loss = loss
    print(f"EPOCH {e}: a = {a}, b = {b}, MSE = {loss}")
    dLoss = mean_squared_error_d(y_train, y_pred)
    da = np.mean(dLoss*X_train)
    db = np.mean(dLoss)
    a -= LEARNING_RATE*da
    b -= LEARNING_RATE*db
    e +=1
```

```python
X_full = np.concatenate((X_train, X_test))
X_full
```

```python
y_full_pred = a * X_full + b
```

```python
plt.scatter(X_train , y_train)
plt.plot(X_train, a*X_train + b)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Train part of the dataset plus prediction")
```

```python
plt.scatter(X_train , y_train)
plt.scatter(X_test , y_test)
plt.plot(X_full, y_full_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Both parts of the dataset plus prediction")
```
